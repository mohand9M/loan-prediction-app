Pourquoi cette étape est-elle importante ?

Les valeurs manquantes dans une colonne peuvent poser des problèmes lors des calculs ou des analyses. Remplacer ces valeurs par une statistique, comme la moyenne, permet de réduire l'impact des données manquantes tout en conservant une cohérence mathématique.

```python
from sklearn.preprocessing import StandardScaler

# Sélectionner les colonnes numériques pour la normalisation
numeric_columns = ['income_annum', 'loan_amount', 'loan_term', 'cibil_score', 
                   'residential_assets_value', 'commercial_assets_value', 
                   'luxury_assets_value', 'bank_asset_value']

scaler = StandardScaler()

# Normaliser les colonnes numériques
dataset_encoded[numeric_columns] = scaler.fit_transform(dataset_encoded[numeric_columns])

# Vérifier les premières lignes du dataset normalisé
print(dataset_encoded.head())
```

Qu'est-ce que train_test_split ?
train_test_split est une fonction utilisée pour :

Diviser les données en deux sous-ensembles :
Jeu d'entraînement (X_train, y_train) : Utilisé pour entraîner le modèle.
Jeu de test (X_test, y_test) : Utilisé pour évaluer les performances du modèle après l'entraînement.
Elle évite ainsi que le modèle s'entraîne et s'évalue sur les mêmes données, ce qui peut entraîner un surajustement (overfitting).

```python
from sklearn.model_selection import train_test_split
```

```python
# Séparer les variables explicatives (X) et la variable cible (y)
X = dataset_encoded.drop(['loan_status_ Rejected'], axis=1)  # Enlevez la colonne 'loan_status_ Rejected'
y = dataset_encoded['loan_status_ Rejected']  # Utilisez 'loan_status_ Rejected' comme cible

# Diviser les données en 70% pour l'entraînement et 30% pour le test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print(f'Ensemble d\'entraînement: {X_train.shape}, Ensemble de test: {X_test.shape}')

```

Ici, vous créez une instance du modèle de régression logistique avec ses paramètres par défaut. La régression logistique est adaptée pour résoudre des tâches de classification binaire, comme prédire si une demande de prêt est approuvée ou rejetée.

X_train : vos variables explicatives.
y_train : la variable cible (le statut du prêt, approuvé ou rejeté).
La méthode fit() entraîne le modèle avec vos données d'entraînement.

X_test : l'ensemble de test avec vos variables explicatives.
predict() utilise le modèle entraîné pour prédire les valeurs de la variable cible.
accuracy_score(y_test, y_pred) : calcule le pourcentage de prédictions correctes.
accuracy * 100 : convertit le score en pourcentage pour une meilleure interprétation.

Qu'est-ce qu'une matrice de confusion ?
C'est une matrice utilisée en classification pour visualiser la performance du modèle. Elle montre :
Vrais positifs (True Positive, TP) : les prédictions correctes pour la classe positive.
Faux positifs (False Positive, FP) : les prédictions incorrectes pour la classe positive.
Vrais négatifs (True Negative, TN) : les prédictions correctes pour la classe négative.
Faux négatifs (False Negative, FN) : les prédictions incorrectes pour la classe négative.

Précision (Precision) : La probabilité qu'une prédiction positive soit correcte.
Rappel (Recall) : La capacité du modèle à identifier tous les vrais positifs.
F1-Score : La moyenne harmonique entre précision et rappel.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

# Initialiser le modèle logistique regression
model = LogisticRegression()

# Entraîner le modèle
model.fit(X_train, y_train)

# Prédire sur l'ensemble de test
y_pred = model.predict(X_test)

# Évaluer la performance du modèle
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy * 100:.2f}%')

# Matrice de confusion
conf_matrix = confusion_matrix(y_test, y_pred)
print(f'Matrice de confusion:\n{conf_matrix}')

# Rapport de classification (Precision, Recall, F1-Score)
class_report = classification_report(y_test, y_pred)
print(f'Rapport de classification:\n{class_report}')
```

```python
from sklearn.ensemble import RandomForestClassifier

# Initialiser le modèle Random Forest
rf_model = RandomForestClassifier(random_state=42)

# Entraîner le modèle
rf_model.fit(X_train, y_train)

# Prédire sur l'ensemble de test
y_pred_rf = rf_model.predict(X_test)

# Évaluer la performance du modèle
accuracy_rf = accuracy_score(y_test, y_pred_rf)
print(f'Accuracy Random Forest: {accuracy_rf * 100:.2f}%')

# Matrice de confusion
conf_matrix_rf = confusion_matrix(y_test, y_pred_rf)
print(f'Matrice de confusion (Random Forest):\n{conf_matrix_rf}')

# Rapport de classification
class_report_rf = classification_report(y_test, y_pred_rf)
print(f'Rapport de classification (Random Forest):\n{class_report_rf}')
```

```python
# =========================
# ÉVALUATION RF & XGB (ROC + Matrices de confusion) — CODE PRÊT À EXÉCUTER
# =========================

import os
import pickle
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import confusion_matrix, roc_curve, auc

# ---------- 0) Charger le dataset ----------
# Cas 1 : tu as déjà une variable 'dataset' définie dans le notebook
# Cas 2 : sinon on essaie de charger le CSV (adapter le chemin si besoin)

if 'dataset' not in globals():
    if os.path.exists('loan_approval_dataset.csv'):
        dataset = pd.read_csv('loan_approval_dataset.csv')
        print("Dataset chargé depuis 'loan_approval_dataset.csv'")
    else:
        raise RuntimeError(
            "Aucun DataFrame 'dataset' trouvé et le fichier 'loan_approval_dataset.csv' est introuvable."
        )

# Nettoyage minimal de la cible
if 'loan_status' not in dataset.columns:
    raise RuntimeError("Colonne cible 'loan_status' introuvable dans le dataset.")

dataset['loan_status'] = dataset['loan_status'].astype(str).str.strip()

# ---------- 1) Séparer X / y + encodage des labels (0/1) ----------
X = dataset.drop('loan_status', axis=1)
y_text = dataset['loan_status']

le = LabelEncoder()
y = le.fit_transform(y_text)   # -> 0/1
classes_text = list(le.classes_)  # ex: ['Approved','Rejected'] (ordre alphabétique)

# ---------- 2) Split ----------
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42, stratify=y
)

# ---------- 3) Utilitaires ----------
def align_features(X_like: pd.DataFrame, model):
    """
    Aligne X_like sur les colonnes vues à l'entraînement par 'model' (feature_names_in_).
    Ajoute les colonnes manquantes remplies à 0 et réordonne les colonnes.
    Si le modèle est un Pipeline, il peut ne pas avoir feature_names_in_, on retourne X_like tel quel.
    """
    if hasattr(model, "feature_names_in_"):
        expected = list(model.feature_names_in_)
        missing = set(expected) - set(X_like.columns)
        for c in missing:
            X_like[c] = 0
        # Supprime les colonnes en trop si besoin
        X_like = X_like[[c for c in expected]]
    return X_like

def positive_proba_column(model, le_target: LabelEncoder):
    """
    Retourne l'indice de la colonne de predict_proba correspondant à la classe positive.
    On tente dans l'ordre :
    - libellé textuel 'Approved' / 'Éligible' (si le modèle a été fit avec strings),
    - label numérique 1 (si le modèle a été fit avec 0/1),
    - sinon on prend la dernière colonne (classe "la plus haute").
    """
    if not hasattr(model, "classes_"):
        return -1
    cls = list(model.classes_)
    # Essai avec des libellés textuels fréquents
    for target_name in ("Approved", "Éligible", "Eligible", "Approuvé"):
        for k, v in enumerate(cls):
            if isinstance(v, str) and v.strip().lower() == target_name.lower():
                return k
    # Essai avec le label numérique 1
    if 1 in cls:
        return cls.index(1)
    # fallback : dernière colonne
    return -1

def plot_conf_matrix(y_true, y_pred, title="Matrice de confusion", cmap="Blues", labels_text=None):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(6, 4))
    sns.heatmap(
        cm, annot=True, fmt="d", cmap=cmap,
        xticklabels=labels_text if labels_text is not None else ["Classe 0", "Classe 1"],
        yticklabels=labels_text if labels_text is not None else ["Classe 0", "Classe 1"]
    )
    plt.xlabel("Prédiction")
    plt.ylabel("Vérité")
    plt.title(title)
    plt.tight_layout()
    plt.show()
    return cm

# ---------- 4) Charger les modèles ----------
rf_model = None
xgb_model = None

if os.path.exists("rf_model.pkl"):
    with open("rf_model.pkl", "rb") as f:
        rf_model = pickle.load(f)
    print("✅ Random Forest chargé: rf_model.pkl")
else:
    print("ℹ️ rf_model.pkl introuvable — la partie RF sera ignorée.")

if os.path.exists("xgb_model.pkl"):
    with open("xgb_model.pkl", "rb") as f:
        xgb_model = pickle.load(f)
    print("✅ XGBoost chargé: xgb_model.pkl")
else:
    print("ℹ️ xgb_model.pkl introuvable — la partie XGB sera ignorée.")

# ---------- 5) Évaluer chaque modèle ----------
roc_curves = []  # pour tracer une ROC comparée si plusieurs modèles dispo

# --- 5.a) Random Forest ---
if rf_model is not None:
    X_test_rf = X_test.copy()
    X_test_rf = align_features(X_test_rf, rf_model)

    # Prédictions classe
    y_pred_rf = rf_model.predict(X_test_rf)

    # Matrice de confusion RF
    _ = plot_conf_matrix(
        y_test, y_pred_rf,
        title="Matrice de confusion — Random Forest",
        cmap="Greens",
        labels_text=classes_text
    )

    # ROC RF
    if hasattr(rf_model, "predict_proba"):
        pos_col_rf = positive_proba_column(rf_model, le)
        y_proba_rf = rf_model.predict_proba(X_test_rf)[:, pos_col_rf]
        fpr_rf, tpr_rf, _ = roc_curve(y_test, y_proba_rf)
        auc_rf = auc(fpr_rf, tpr_rf)

        plt.figure(figsize=(8, 6))
        plt.plot(fpr_rf, tpr_rf, lw=2, label=f"Random Forest (AUC = {auc_rf:.2f})")
        plt.plot([0, 1], [0, 1], linestyle="--", color="gray")
        plt.xlim([0, 1]); plt.ylim([0, 1.05])
        plt.xlabel("Taux de faux positifs (FPR)")
        plt.ylabel("Taux de vrais positifs (TPR)")
        plt.title("Courbe ROC — Random Forest")
        plt.legend(loc="lower right")
        plt.tight_layout()
        plt.show()

        roc_curves.append(("Random Forest", fpr_rf, tpr_rf, auc_rf))

# --- 5.b) XGBoost ---
if xgb_model is not None:
    X_test_xgb = X_test.copy()
    X_test_xgb = align_features(X_test_xgb, xgb_model)

    # Prédictions classe
    y_pred_xgb = xgb_model.predict(X_test_xgb)

    # Matrice de confusion XGB
    _ = plot_conf_matrix(
        y_test, y_pred_xgb,
        title="Matrice de confusion — XGBoost",
        cmap="Blues",
        labels_text=classes_text
    )

    # ROC XGB
    if hasattr(xgb_model, "predict_proba"):
        pos_col_xgb = positive_proba_column(xgb_model, le)
        y_proba_xgb = xgb_model.predict_proba(X_test_xgb)[:, pos_col_xgb]
        fpr_xgb, tpr_xgb, _ = roc_curve(y_test, y_proba_xgb)
        auc_xgb = auc(fpr_xgb, tpr_xgb)

        plt.figure(figsize=(8, 6))
        plt.plot(fpr_xgb, tpr_xgb, lw=2, label=f"XGBoost (AUC = {auc_xgb:.2f})")
        plt.plot([0, 1], [0, 1], linestyle="--", color="gray")
        plt.xlim([0, 1]); plt.ylim([0, 1.05])
        plt.xlabel("Taux de faux positifs (FPR)")
        plt.ylabel("Taux de vrais positifs (TPR)")
        plt.title("Courbe ROC — XGBoost")
        plt.legend(loc="lower right")
        plt.tight_layout()
        plt.show()

        roc_curves.append(("XGBoost", fpr_xgb, tpr_xgb, auc_xgb))

# ---------- 6) Courbe ROC comparée (si ≥ 2 modèles) ----------
if len(roc_curves) >= 2:
    plt.figure(figsize=(8, 6))
    for name, fpr, tpr, val_auc in roc_curves:
        plt.plot(fpr, tpr, lw=2, label=f"{name} (AUC = {val_auc:.2f})")
    plt.plot([0, 1], [0, 1], linestyle="--", color="gray")
    plt.xlim([0, 1]); plt.ylim([0, 1.05])
    plt.xlabel("Taux de faux positifs (FPR)")
    plt.ylabel("Taux de vrais positifs (TPR)")
    plt.title("Courbes ROC — Comparaison des modèles")
    plt.legend(loc="lower right")
    plt.tight_layout()
    plt.show()
```

1. score_cibil (0.808006)
Description : Le score CIBIL (ou score de crédit) est une mesure de la fiabilité financière d'une personne. Il est calculé en fonction des antécédents de crédit (remboursement de prêts, paiements de factures, etc.).
Importance : Très élevé (0.808) : C'est la variable la plus importante pour décider si un prêt est approuvé ou non. Un bon score CIBIL augmente les chances d'obtenir un prêt.

2. durée_du_prêt (loan_term, 0.054689)
Description : Représente la durée du prêt (en mois ou en années). Une durée plus longue peut réduire les mensualités mais augmenter les intérêts totaux.
Importance : Moyenne (0.054). Cette caractéristique influence modérément la décision, car les banques préfèrent des durées de prêt qui correspondent à leur politique de risque.

3. montant_du_prêt (loan_amount, 0.026244)
Description : Montant total demandé par l'emprunteur pour le prêt.
Importance : Moyenne (0.026). Les banques analysent le montant demandé pour s'assurer qu'il est réaliste et en adéquation avec la capacité de remboursement de l'emprunteur.

4. id_prêt (loan_id, 0.018538)
Description : Identifiant unique du prêt. Il est utilisé pour différencier les prêts mais n'a pas de lien direct avec l'approbation.
Importance : Faible (0.018). Utilisé uniquement pour le suivi des données, pas pour la prise de décision.

5. valeur_des_actifs_résidentiels (residential_assets_value, 0.017418)
Description : Valeur totale des propriétés résidentielles détenues par l'emprunteur. Par exemple, des maisons ou des appartements.
Importance : Moyenne (0.017). Une plus grande valeur des actifs rassure les banques sur la stabilité financière de l'emprunteur.

6. valeur_des_actifs_de_luxe (luxury_assets_value, 0.016573)
Description : Valeur des biens de luxe possédés par l'emprunteur (comme des voitures de luxe, des bijoux, etc.).
Importance : Moyenne-faible (0.016). Cela montre un potentiel de richesse, mais ce n'est pas un facteur clé pour l'approbation des prêts.

7. revenu_annuel (income_annum, 0.016299)
Description : Revenus annuels de l'emprunteur. Cela inclut les salaires, les revenus d'entreprise ou d'autres sources.
Importance : Moyenne-faible (0.016). Un revenu élevé est souvent requis pour des prêts importants, mais ce n'est pas le seul facteur pris en compte.

8. valeur_des_actifs_commerciaux (commercial_assets_value, 0.015225)
Description : Valeur des actifs commerciaux détenus, comme des bureaux, des magasins ou des entreprises.
Importance : Moyenne-faible (0.015). Cela montre une stabilité financière supplémentaire, mais son importance est limitée.

9. valeur_des_actifs_bancaires (bank_asset_value, 0.014909)
Description : Valeur des actifs financiers détenus en banque, comme les comptes épargne, les investissements, etc.
Importance : Moyenne-faible (0.015). Plus cette valeur est élevée, plus l'emprunteur est perçu comme solvable.

10. nombre_de_personnes_à_charge (no_of_dependents, 0.007752)
Description : Nombre de personnes dépendant financièrement de l'emprunteur (enfants, parents, etc.).
Importance : Faible (0.007). Plus il y a de personnes à charge, plus la capacité de remboursement peut être limitée, mais cet effet est minime.

11. niveau_d’éducation (education, 0.002219)
Description : Niveau d'éducation atteint par l'emprunteur (ex. : secondaire, universitaire, etc.).
Importance : Très faible (0.002). Bien qu'un niveau d'éducation élevé puisse indiquer une meilleure employabilité, son impact est faible sur l'approbation du prêt.

12. travailleur_indépendant (self_employed, 0.002128)
Description : Statut d'emploi de l'emprunteur. Indique si l'emprunteur est salarié ou travaille à son compte.
Importance : Très faible (0.002). Les travailleurs indépendants peuvent être perçus comme plus risqués, mais cet effet est marginal.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from xgboost import XGBClassifier
from sklearn.metrics import roc_curve, auc

# Charger le dataset
df = pd.read_csv("loan_approval_dataset.csv")
df.columns = df.columns.str.strip()

# Encoder les colonnes catégorielles
for col in ["education", "self_employed", "loan_status"]:
    df[col] = LabelEncoder().fit_transform(df[col].astype(str))

# Variables indépendantes et cible
X = df.drop("loan_status", axis=1)
y = df["loan_status"]

# Normalisation
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split train/test
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.3, random_state=42, stratify=y
)

# Définir les modèles
models = {
    "Régression logistique": LogisticRegression(max_iter=1000),
    "Arbre de décision": DecisionTreeClassifier(max_depth=5, random_state=42),
    "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42),
    "SVM": SVC(probability=True, kernel="rbf", random_state=42),
    "XGBoost": XGBClassifier(use_label_encoder=False, eval_metric="logloss", random_state=42)
}

# Tracer les courbes ROC
plt.figure(figsize=(8,6))

for name, model in models.items():
    model.fit(X_train, y_train)
    y_proba = model.predict_proba(X_test)[:,1]  # probabilité de classe positive
    fpr, tpr, _ = roc_curve(y_test, y_proba)
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, label=f"{name} (AUC = {roc_auc:.2f})")

# Courbe aléatoire (baseline)
plt.plot([0,1],[0,1],'k--', label="Aléatoire")

plt.title("Figure III.2 – Courbe ROC comparée des modèles")
plt.xlabel("Taux de faux positifs (FPR)")
plt.ylabel("Taux de vrais positifs (TPR)")
plt.legend(loc="lower right")
plt.grid(True)
plt.tight_layout()

# Sauvegarde (image utilisable dans ton mémoire)
plt.savefig("roc_comparaison.png", dpi=300)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestClassifier

# Charger le dataset
df = pd.read_csv("loan_approval_dataset.csv")
df.columns = df.columns.str.strip()

# Encoder les colonnes catégorielles
for col in ["education", "self_employed", "loan_status"]:
    df[col] = LabelEncoder().fit_transform(df[col].astype(str))

# Variables et cible
X = df.drop("loan_status", axis=1)
y = df["loan_status"]

# Standardisation
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split train/test
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.3, random_state=42, stratify=y
)

# Entraînement Random Forest
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)

# Importance des variables
importances = rf.feature_importances_
features = X.columns

# Bar chart
plt.figure(figsize=(8,6))
plt.barh(features, importances, color="skyblue")
plt.xlabel("Importance")
plt.title("Figure III.3 – Importance des variables (Random Forest)")
plt.gca().invert_yaxis()  # plus lisible
plt.tight_layout()

# Sauvegarde en image
plt.savefig("feature_importance_rf.png", dpi=300)

plt.show()
```
